# Feature Engineering Using Date columns

```python
df['START_DATE']=df['START_DATE'].str.replace('/','-')
df['END_DATE']=df['END_DATE'].str.replace('/','-')
```

```python
df.info()
```

```python
df['START_DATE']=pd.to_datetime(df['START_DATE'])
df['END_DATE']=pd.to_datetime(df['END_DATE'])
```

```python
df.info()
```

```python
#splitting the START_DATE to date and time column and then converting the time intofour different categories (M,A,E,N)

df['DATE']=pd.DatetimeIndex(df['START_DATE']).date
df['TIME']=pd.DatetimeIndex(df['START_DATE']).hour
```

```python
df.head()
```

```python
df.info()
```

```python
out=[]
for i in df['TIME']:
  if 0<=i<=11:
    out.append('Morning')
  elif 12<=i<=16:
    out.append('Afternoon')
  elif 17<=i<=19:
    out.append('Evening')
  elif 20<=i<=24:
    out.append('Night')

df['DAY-NIGHT']=out
df.head()
```

```python
#creating Month column

# df['Month']=[i.month for i in df['START_DATE']]
# df['Month']=df['Month'].map({1:'JAN',2:'FEB',3:'MAR',4:'APR',5:'MAY',6:'JUN',7:'JUL',8:'AUG',9:'SEP',10:'OCT',11:'NOV',12:'DEC'})

df['Month']=df['START_DATE'].dt.strftime('%b')

df.head()
```

```python
#creating day column
df['Day']=df['START_DATE'].dt.strftime('%a')
```

# Count by different columns

```python
b=sns.countplot(x='CATEGORY',data=df,hue='CATEGORY',palette='dark:black')
def data_label(arg):
  for i in arg.containers:
    arg.bar_label(i,label_type='center',color='white')
data_label(b)
```
